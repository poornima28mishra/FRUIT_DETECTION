```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
cnn = tf.keras.models.load_model("C:/Users/Poorn/Documents/fruit classification/my_model.keras")
```

```python
import cv2
image_path = "D:/PROJECT/Fruit_dataset/test/carrot/Image_8.jpg"
img = cv2.imread(image_path)
plt.imshow(img)
plt.title("test img")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64,3))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr =np.array([input_arr])
prediction = cnn.predict(input_arr)
```

```python
print(prediction[0])
print(max(prediction[0]))
```

```python
test_set= tf.keras.utils.image_dataset_from_directory("D:/PROJECT/Fruit_dataset/test",
            labels = 'inferred',
            label_mode='categorical',
            class_names=None,
            color_mode='rgb',
            batch_size=32,
            image_size= (64,64),
            shuffle=True,
            seed=None,
            validation_split=None,
            subset =None,
            interpolation='bilinear',
            follow_links=False,
            crop_to_aspect_ratio=False
        )
```
